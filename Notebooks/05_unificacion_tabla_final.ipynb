```python
import pandas as pd
```

```python
df_clientes = pd.read_csv("../data/clean/clientes_limpios.csv")
```

```python
df_productos = pd.read_csv("../data/clean/productos_limpios.csv")
```

```python
df_ventas = pd.read_csv("../data/clean/ventas_limpios.csv")
```

```python
df_detalle_ventas = pd.read_csv("../data/clean/detalle_ventas_limpios.csv")
```

```python
print("CLIENTES")
display(df_clientes.head())

print("PRODUCTOS")
display(df_productos.head())

print("VENTAS")
display(df_ventas.head())

print("DETALLE VENTAS")
display(df_detalle_ventas.head())
```

```python
df_vc = df_ventas.merge(df_clientes, on="id_cliente", how="left")
df_vc.head()
```

```python
df_vcd = df_vc.merge(df_detalle_ventas, on="id_venta", how="left")
df_vcd.head()
```

```python
df_final = df_vcd.merge(df_productos, on="id_producto", how="left")
df_final.head()
```

```python
df_final.columns
```

```python
df_unificado = df_final[[
    'id_venta',
    'fecha',
    'id_cliente',
    'nombre_cliente_y',
    'email_y',
    'ciudad',
    'fecha_alta',
    'medio_pago',
    'id_producto',
    'nombre_producto_y',
    'categoria',
    'cantidad',
    'precio_unitario_y',
    'importe'
]]
```

```python
df_unificado = df_unificado.rename(columns={
    'nombre_cliente_y': 'nombre_cliente',
    'email_y': 'email',
    'nombre_producto_y': 'nombre_producto',
    'precio_unitario_y': 'precio_unitario'
})
```

```python
# código para cargar las correcciones

from collections import OrderedDict

correccion_categoria = OrderedDict({
    'lavandina': 'limpieza',
    'trapo': 'limpieza',
    'limpiavidrios': 'limpieza',
    'desengrasante': 'limpieza',
    'detergente': 'limpieza',
    'jabón': 'limpieza',
    'papel higiénico': 'limpieza',
    'toalla': 'limpieza',
    'hilo dental': 'limpieza',
    'cepillo': 'limpieza',
    'crema dental': 'limpieza',
    'desodorante': 'limpieza',
    'shampoo': 'limpieza',
    'mascarilla': 'limpieza',
    'servilleta': 'limpieza',
})

def corregir_categoria(nombre_producto):
    nombre = nombre_producto.lower()
    
    for clave, categoria in correccion_categoria.items():
        if clave in nombre:
            return categoria
    
    return 'alimentos'

# Aplicar la función
df_unificado['categoria'] = df_unificado['nombre_producto'].apply(corregir_categoria)

# Ver resultados
print(df_unificado['categoria'].value_counts())

# Opcional: Ver listas de productos
for categoria, grupo in df_unificado.groupby('categoria'):
    print(f"\n{categoria.capitalize()}:")
    for producto in sorted(grupo['nombre_producto'].unique()):
        print(f"- {producto}")
```

```python
# Ver total de filas en df_unificado
print(f"Total de filas en df_unificado: {len(df_unificado)}")

# Ver productos únicos por categoría
print("Productos únicos en alimentos:", len(df_unificado[df_unificado['categoria'] == 'alimentos']['nombre_producto'].unique()))
print("Productos únicos en limpieza:", len(df_unificado[df_unificado['categoria'] == 'limpieza']['nombre_producto'].unique()))

# Si quieres eliminar duplicados basados en 'nombre_producto' (manteniendo la primera ocurrencia)
df_sin_duplicados = df_unificado.drop_duplicates(subset=['nombre_producto'])
print(f"Después de eliminar duplicados: {len(df_sin_duplicados)} filas")
print(df_sin_duplicados['categoria'].value_counts())
```
